```python
import glob
import re

import pandas as pd
import spotify_mexico.data.load as load
```

### Load data

```python
data_raw = load.data_raw
data_processed = load.data_processed
```

### Read files as dataframes

```python
# All files ending with .csv
paths = glob.glob(f'{data_raw}/*.csv')
dfs = []
# Read files as dataframes
for i in paths:
    df = load.load_file(file=f'{i}')
    dfs.append(df)
```

### Process data

```python
for i in range(len(paths)):
    date = re.findall(r'weekly-(\d*-\d*-\d*)', paths[i])
    year = re.findall(r'weekly-(\d*)', paths[i])
    month = re.findall(r'weekly-\d*-(\d*)', paths[i])
    days = re.findall(r'weekly-\d*-\d*-(\d*)', paths[i])
    day = int(days[0])
    if day <= 7:
        week = 1
    if day > 7 and day <= 14:
        week = 2
    if day > 14 and day <= 21:
        week = 3
    if day > 21 and day <= 28:
        week = 4
    if day > 28:
        week = 5
    dfs[i]['date'] = date[0]
    dfs[i]['year'] = year[0]
    dfs[i]['month'] = month[0]
    dfs[i]['week'] = week
```

### Concatenate dataframes

```python
df = pd.concat(dfs)
df
```

### Save dataframe

```python
df.to_csv(f'{data_processed}/spotify_mexico.csv', index=False)
```
